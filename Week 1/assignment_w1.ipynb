**Consider the following function $$s (t) = 2 e^{-t} \cos(2\pi t).$$**

**Using sympy (or numpy or scipy), do the following tasks**

- Define the function.

```python
import sympy as sp
import matplotlib.pyplot as plt
from matplotlib import style
style.use('bmh')

import warnings
warnings.filterwarnings('ignore')

sp.init_printing()

plt.rcParams['figure.figsize'] = 16,6   

x = sp.symbols('x') 
t = sp.symbols('t')

s = 2*sp.exp(-t)*sp.cos(2*sp.pi*t)
s
```

- $\mathrm{d}s/\mathrm{d}t$

```python
s_der = sp.diff(s, t)
s_der
```

- $\int s(t) \mathrm{d}t$

```python
s_int = sp.integrate(s, t)
s_int
```

- Plot the three signals together.

```python
plots = sp.plot(s, s_der, s_int, show = False)
plots[0].line_color = 'r'
plots[1].line_color = 'g'
plots[2].line_color = 'b'
plots.show()
```

- Fourier tranform of $S(w) = \mathcal{F} \left\{s(t) u(t)\right\},$ where $u(t)$ is the step function.

```python
omega = sp.symbols('omega')
#ft_t is Fourier transform of S(w)
ft_s = sp.fourier_transform(s * sp.Heaviside(t), t, omega)
ft_s
```

- Plot the real part of $S(w), \; -5 \leq w \leq 5$

```python
sp.plot(sp.re(ft_s), (omega, -5, 5))
```

- Plot the imaginary part of $S(w), \; -5 \leq w \leq 5$

```python
sp.plot(sp.im(ft_s), (omega, -5, 5))
```

- Plot the absolute value of $S(w), \; -5 \leq w \leq 5$

```python
sp.plot(sp.Abs(ft_s), (omega, -5, 5))
```

- Plot the phase of $S(w), \; -5 \leq w \leq 5$

```python
sp.plot(sp.arg(ft_s), (omega, -5, 5))
```
